```python
# Function to read data from ARFF file and store it in a matrix
def read_arff(file_path):
    data = []
    with open(file_path, 'r') as file:
        data_started = False
        for line in file:
            line = line.strip()
            if data_started:
                if line != '' and not line.startswith('%'):
                    row = line.split(',')
                    data.append(row)
            elif line.lower().startswith('@data'):
                data_started = True
    return data

# Function to store data from multiple ARFF files into a matrix
def store_arff_data(arff_files):
    all_data = []
    for file in arff_files:
        data = read_arff(file)
        all_data.extend(data)
    return all_data

arff_files = [
    '2017 Q1.arff', '2017 Q2.arff', '2017 Q3.arff', '2017 Q4.arff', '2017.arff',
    '2018 Q1.arff', '2018 Q2.arff', '2018 Q3.arff', '2018 Q4.arff', '2018.arff',
    '2019 Q1.arff', '2019 Q2.arff', '2019 Q3.arff', '2019 Q4.arff', '2019.arff',
    '2020 Q1.arff', '2020 Q2.arff', '2020 Q3.arff', '2020 Q4.arff', '2020.arff',
    '2021 Q1.arff'
]

# Store data from ARFF files into a matrix
all_arff_data = store_arff_data(arff_files)

# After storing data from ARFF files into a matrix
print("Total number of rows in Stored ARFF Data:", len(all_arff_data))

# Print the first 10 rows of the stored data (matrix)
print("Stored ARFF Data (First 10 rows):")
for i, row in enumerate(all_arff_data[:10]):
    print(f"Row {i + 1}: {row}")
```

```python
def convert_to_numeric(data):
    converted_data = []

    for row in data:
        converted_row = []
        for item in row:
            try:
                # Try converting the entry to float
                converted_item = float(item)
            except ValueError:
                # If it's not convertible to float, check for numeric types (int, float)
                if isinstance(item, (int, float)):
                    converted_item = item
                else:
                    # If it's a string, convert it to its numerical equivalent
                    try:
                        converted_item = float(item)
                    except ValueError:
                        # If it's not a convertible string, replace with a default numeric value
                        converted_item = 1.0  # You can set any default value here instead of 0.0
            converted_row.append(converted_item)
        converted_data.append(converted_row)
    return converted_data

# Example: Convert non-numeric entries in all_arff_data to numeric equivalents
numeric_data = convert_to_numeric(all_arff_data)

# Display the first 10 rows of the converted numeric data
print("Converted Numeric Data (First 10 rows):")
for i, row in enumerate(numeric_data[:10]):
    print(f"Row {i + 1}: {row}")
```

**Performing PCA**

```python
# Function to calculate mean of columns
# Function to check if a value is numeric
def is_numeric(value):
    try:
        float(value)
        return True
    except ValueError:
        return False

# Function to calculate mean of columns (excluding non-numeric values)
def calculate_mean(data):
    num_columns = len(data[0])
    num_rows = len(data)
    means = [0] * num_columns

    for i in range(num_columns):
        column_sum = 0
        count = 0
        for j in range(num_rows):
            if is_numeric(data[j][i]):
                column_sum += float(data[j][i])
                count += 1
        if count > 0:
            means[i] = column_sum / count

    return means

# Function to subtract mean from data (excluding non-numeric values)
def mean_subtraction(data, mean):
    num_rows = len(data)
    num_columns = len(data[0])
    centered_data = []

    for i in range(num_rows):
        centered_row = []
        for j in range(num_columns):
            if is_numeric(data[i][j]):
                centered_row.append(float(data[i][j]) - mean[j])
        if len(centered_row) > 0:
            centered_data.append(centered_row)

    return centered_data



# Function to perform matrix transposition
def transpose_matrix(matrix):
    return [[matrix[j][i] for j in range(len(matrix))] for i in range(len(matrix[0]))]

# Function to calculate covariance matrix
def calculate_covariance_matrix(data):
    num_samples = len(data)
    num_features = len(data[0])
    covariance_matrix = [[0] * num_features for _ in range(num_features)]

    for i in range(num_features):
        for j in range(i, num_features):
            sum_val = 0
            for k in range(num_samples):
                sum_val += data[k][i] * data[k][j]
            covariance_matrix[i][j] = sum_val / (num_samples - 1)
            covariance_matrix[j][i] = covariance_matrix[i][j]

    return covariance_matrix

# Function to perform eigenvalue decomposition
def eigen_decomposition(matrix):
    eigenvalues = []
    eigenvectors = []

    # Initial approximation for eigenvectors
    for i in range(len(matrix)):
        eigenvector = [0] * len(matrix)
        eigenvector[i] = 1
        eigenvectors.append(eigenvector)

    for _ in range(100):  # Number of iterations for convergence
        new_eigenvectors = []

        for i in range(len(matrix)):
            new_eigenvector = [0] * len(matrix)
            for j in range(len(matrix)):
                val = 0
                for k in range(len(matrix)):
                    val += matrix[i][k] * eigenvectors[k][j]
                new_eigenvector[j] = val

            # Normalize the eigenvector
            norm = max(new_eigenvector)
            new_eigenvector = [x / norm for x in new_eigenvector]
            new_eigenvectors.append(new_eigenvector)

        eigenvectors = new_eigenvectors

        # Calculate eigenvalues (using Rayleigh quotient)
        eigenvalues = [sum(matrix[i][j] * eigenvectors[i][j] for i in range(len(matrix))) for j in range(len(matrix))]

    return eigenvalues, eigenvectors



# Function to perform PCA
def principal_component_analysis(data, num_components):
    mean = calculate_mean(data)
    centered_data = mean_subtraction(data, mean)
    covariance_matrix = calculate_covariance_matrix(centered_data)
    eigenvalues, eigenvectors = eigen_decomposition(covariance_matrix)

    # Sorting eigenvalues and corresponding eigenvectors
    eigen_pairs = [(eigenvalues[i], eigenvectors[i]) for i in range(len(eigenvalues))]
    eigen_pairs.sort(reverse=True)

    sorted_eigenvectors = [pair[1] for pair in eigen_pairs]

    # Selecting top 'num_components' eigenvectors
    selected_eigenvectors = sorted_eigenvectors[:num_components]

    # Projecting data onto the new space formed by the selected eigenvectors
    projected_data = [[sum(centered_data[i][j] * selected_eigenvectors[k][j] for j in range(len(centered_data[0])))
                      for k in range(num_components)] for i in range(len(centered_data))]

    return projected_data

# Perform PCA on data with 2 principal components
num_principal_components = 2
pca_result = principal_component_analysis(all_arff_data, num_principal_components)

# Print the results of PCA
print(pca_result)
```

**Performing SVD**

```python
# Helper function to compute the mean of a list of numbers
def mean(values):
    return sum(values) / len(values)

# Helper function to compute the dot product of two lists/vectors
def dot_product(a, b):
    return sum(x * y for x, y in zip(a, b) if isinstance(x, (int, float)) and isinstance(y, (int, float)))

# Function to perform matrix multiplication
def matrix_multiply(matrix_a, matrix_b):
    result = []
    for row in matrix_a:
        new_row = []
        for col_index in range(len(matrix_b[0])):
            col = [matrix_b[i][col_index] for i in range(len(matrix_b))]
            product = dot_product(row, col)
            new_row.append(product)
        result.append(new_row)
    return result

# Function to transpose a matrix
def transpose(matrix):
    return [list(row) for row in zip(*matrix)]

# Check and convert strings to floats
def convert_to_float(value):
    try:
        if value.isdigit() or (value[0] == '-' and value[1:].isdigit()):
            return float(value)
        else:
            return value
    except ValueError:
        return value  # Preserve non-numeric values

# Function to convert all data to floats or replace non-numeric values
def convert_data_to_numeric(data):
    return [[convert_to_float(value) for value in row] for row in data]

# Function to calculate factorial
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        result = 1
        for i in range(2, n + 1):
            result *= i
        return result

# Function to compute cosine using Taylor series expansion
def custom_cos(theta, terms=10):
    theta = theta % (2 * 3.14159)  # Ensure angle is within range (0, 2*pi)
    result = 0
    for n in range(terms):
        sign = (-1) ** n
        term = ((theta ** (2 * n)) / factorial(2 * n))
        result += sign * term
    return result

# Function to compute sine using Taylor series expansion
def custom_sin(theta, terms=10):
    theta = theta % (2 * 3.14159)  # Ensure angle is within range (0, 2*pi)
    result = 0
    for n in range(terms):
        sign = (-1) ** n
        term = ((theta ** ((2 * n) + 1)) / factorial((2 * n) + 1))
        result += sign * term

    # Adjust sign for negative angles
    if theta > 3.14159:  # If theta is in the range (pi, 2*pi), then it's a negative angle
        result = -result
    
    return result


# Function to perform Singular Value Decomposition (SVD)
def svd(X, tolerance=1e-10):
    # Convert data to numeric
    X = convert_data_to_numeric(X)
    # Initialize U, Sigma, and V matrices
    num_rows, num_cols = len(X), len(X[0])
    U = [[0 for _ in range(num_rows)] for _ in range(num_rows)]
    Sigma = [0 for _ in range(min(num_rows, num_cols))]
    V = [[0 for _ in range(num_cols)] for _ in range(num_cols)]
    
    # Implement the Golub-Kahan SVD algorithm (a basic version)
    B = matrix_multiply(transpose(X), X)
    V = [[1 if i == j else 0 for j in range(num_cols)] for i in range(num_cols)]

    while True:
        # Finding the largest off-diagonal element
        max_off_diag = max(B[i][j] for i in range(num_cols) for j in range(i+1, num_cols))
        
        if max_off_diag < tolerance:
            break

        for i in range(num_cols - 1):
            for j in range(i+1, num_cols):
                if B[i][j] == max_off_diag:
                    p = i
                    q = j

        # Compute rotation angle
        if B[p][p] == B[q][q]:
            theta = 0.25 * 3.14159 if B[p][q] > 0 else -0.25 * 3.14159
        else:
            phi = 0.5 * (2 * B[p][q] / (B[p][p] - B[q][q]))
            theta = phi + 3.14159/2 if phi >= 0 else phi - 3.14159/2

        # Create Givens rotation matrix
        G = [[1 if i == j else 0 for j in range(num_cols)] for i in range(num_cols)]
        G[p][p] = custom_cos(theta)
        G[p][q] = custom_sin(theta)
        G[q][p] = custom_sin(theta)
        G[q][q] = custom_cos(theta)

        # Update matrices
        B = matrix_multiply(transpose(G), B)
        B = matrix_multiply(B, G)
        V = matrix_multiply(V, G)

    # Compute singular values and update U matrix
    for i in range(min(num_rows, num_cols)):
        Sigma[i] = B[i][i] ** 0.5
        if B[i][i] >= 0:
            U[i][i] = 1
        else:
            U[i][i] = -1

    return U, Sigma, V

U, Sigma, V = svd(all_arff_data)

# Perform SVD
U, Sigma, V = svd(all_arff_data)
print("\nSVD Result:")
print("U:")
print(U)
print("\nSingular Values:")
print(Sigma)
print("\nV:")
print(V)
```
